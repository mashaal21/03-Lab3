```python
#Increase the size of the image.
import cv2 
import matplotlib.pyplot as plt


img= cv2.imread('cameraman.jpg', 0)

scaled =cv2.resize(img , None, fx=2 , fy=2 , interpolation=cv2.INTER_CUBIC)

plt.imshow(scaled, cmap='gray')
plt.title('Scaled Image (2x)')
plt.axis('off')
plt.show()
```

```python
#Rotate the image with 120 degrees
import cv2 
import matplotlib.pyplot as plt
 
img = cv2.imread('cameraman.jpg',0)

(h,w) = img.shape

center = (w//2 , h//2)

M = cv2.getRotationMatrix2D(center , 120 , 1.0)

rotated = cv2.warpAffine(img , M, (w,h))

plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.show()


plt.imshow(rotated, cmap='gray')
plt.title('Rotated (120°)')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('cameraman.jpg', 0)

h, w = img.shape

# Shear في اتجاه X
shear_factor = 0.5

M = np.float32([
    [1, shear_factor, 0],
    [0, 1, 0]
])

new_w = int(w + shear_factor * h)

sheared = cv2.warpAffine(img, M, (new_w, h))

plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.show()

plt.imshow(sheared, cmap='gray')
plt.title('Sheared Image')
plt.axis('off')
plt.show()
```

Task2: Use any image processing library and apply the following intensity
transformations over the image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('cameraman.jpg', 0)
```

```python
#Negative
negative = 255 - img

plt.imshow(negative, cmap='gray')
plt.title('Negative')
plt.axis('off')
plt.show()
```

```python
img = cv2.imread('cameraman.jpg', 0)



img_float = img.astype(np.float32)

c = 255 / np.log(1 + np.max(img_float))
log_img = c * np.log(1 + img_float)

log_img = np.uint8(log_img)

plt.imshow(log_img, cmap='gray')
plt.title('Log Transformation')
plt.axis('off')
plt.show()
```

```python
img = cv2.imread('cameraman.jpg', 0)

gamma = 0.2   
c = 1

power = c * np.power(img / 255.0, gamma)
power = np.array(power * 255, dtype=np.uint8)

plt.imshow(power, cmap='gray')
plt.title('Power Law (Gamma=0.5)')
plt.axis('off')
plt.show()
```
